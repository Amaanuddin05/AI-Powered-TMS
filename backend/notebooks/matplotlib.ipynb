```python
goal_types=["penalty","freekick","field"]
goals=[20,11,69]
clr=["red","yellow","orange"]
plt.title("Goal Distribution",pad=30)

plt.pie(goals,labels=goal_types,autopct="%.2f%%",shadow=True,explode=(0,0.2,0),colors=clr)

plt.show()
```
